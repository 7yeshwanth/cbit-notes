````r
treated_data <- prepare(treat_plan, data)
head(treated_data)
````

````r
mn = mean(treated_data$Age)
mn
````

````r
sda = sd(treated_data$Age)
sda
````

````r
mn + c(-sda, sda)
````

````r
treated_data$sage <- (treated_data$Age - mn)/sda
head(treated_data)
````

````r
treated_data[abs(treated_data["Age"] - mn) < sda, ]
````

## **1. `select()` Function in _dplyr_**

The **`select()`** function is used to choose specific columns from a dataframe. It can be used to:
- **Select specific columns.**
- **Drop unnecessary columns.**
- **Rename or reorder columns.**
- **Utilize helper functions** to match column names based on patterns.



## **Key Features**

1. **Select Specific Columns**:  
   *You can choose one or more columns by name.*
   
2. **Drop Columns**:  
   *Use the `-` sign to exclude certain columns.*
   
3. **Rename Columns**:  
   *Rename columns directly within the `select()` function.*
   
4. **Use Helper Functions**:
   - **`starts_with()`**: *Selects columns that start with a prefix.*
   - **`ends_with()`**: *Selects columns that end with a suffix.*
   - **`contains()`**: *Selects columns containing a substring.*
   - **`matches()`**: *Selects columns matching a regular expression.*





## **Examples**

### **1. Select Specific Columns**
```r
mtcars %>% select(mpg, hp, cyl)
```





### **2. Exclude Specific Columns**
```r
mtcars %>% select(-hp, -cyl)
```



### **3. Rename Columns While Selecting**
```r
mtcars %>% select(mileage = mpg, horsepower = hp, cylinders = cyl)
```



### **4. Select Columns with Helper Functions**
```r
# Select columns that start with 'd'
mtcars %>% select(starts_with("d"))

# Select columns that contain the letter 'a'
mtcars %>% select(contains("a"))
```



### **5. Reorder Columns**
```r
mtcars %>% select(cyl, mpg, hp)
```



### **6. Select Columns by Index**
```r
mtcars %>% select(1:3)
```





### **7. Combine `select()` with Other `dplyr` Functions**
```r
mtcars %>% mutate(weight_kg = wt * 453.592) %>% select(mpg, cyl, weight_kg)
```
