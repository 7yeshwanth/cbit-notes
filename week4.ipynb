## ***3. Analysis of XML Files***

### ***Exercise 3.1: Convert the `contacts.xml` file to a data frame in R with columns for id, name, and email***
```r
# Load the XML file
contacts_xml <- xmlParse("contacts.xml")

# Extract data
ids <- xpathSApply(contacts_xml, "//contact", xmlGetAttr, "id")
names <- xpathSApply(contacts_xml, "//name", xmlValue)
emails <- xpathSApply(contacts_xml, "//email", xmlValue)

# Create data frame
contacts_df <- data.frame(
    id = ids,
    name = names,
    email = emails,
    stringsAsFactors = FALSE
)

contacts_df
```



### ***Exercise 3.2: Convert the `products.xml` file to a data frame in R with columns for id, name, and price***
```r
# Load the XML file
products_xml <- xmlParse("products.xml")

# Extract data
ids <- xpathSApply(products_xml, "//product", xmlGetAttr, "id")
names <- xpathSApply(products_xml, "//name", xmlValue)
prices <- xpathSApply(products_xml, "//price", xmlValue)

# Create data frame
products_df <- data.frame(
    id = ids,
    name = names,
    price = as.numeric(prices),
    stringsAsFactors = FALSE
)

products_df
```
